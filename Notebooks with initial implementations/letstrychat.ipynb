```julia
using MPSKit

function generate_entangled_mps(L, D, physical_size)
    tensor_list = []
    
    for i in 1:Int(L)
        if i < Int(L/2)
            dims = (D*physical_size, D)
            QL = TensorMap(randisometry(dims), ℝ^(D) ⊗ ℝ^(physical_size), ℝ^D)
            push!(tensor_list, QL)
        end
        function generate_entangled_mps(L, D, physical_size)
            tensor_list = []
            
            for i in 1:Int(L)
                if i < Int(L/2)
                    dims = (D*physical_size, D)
                    QL = TensorMap(randisometry(dims), ℝ^(D) ⊗ ℝ^(physical_size), ℝ^D)
                    push!(tensor_list, QL)
                end
        
                if i == Int(L/2)
                    QL = TensorMap(randisometry((D*physical_size, D)), ℝ^(D) ⊗ ℝ^(physical_size), ℝ^D)
                    QR = transpose(TensorMap(randisometry((D*physical_size, D)), ℝ^(physical_size) ⊗ ℝ^(D), ℝ^(D)))
                    entanglement_spectrum = TensorMap(sqrt(diagm([1/5 for i in 1:5])), ℝ^D, ℝ^D)
        
                    QL_mid = QL * entanglement_spectrum
                    QR_mid = entanglement_spectrum * QR
        
                    push!(tensor_list, QL_mid)
                    push!(tensor_list, QR_mid)
                end
        
                if i > Int(L/2)+1
                    QR = transpose(TensorMap(randisometry((D*physical_size, D)), ℝ^(physical_size) ⊗ ℝ^(D), ℝ^(D)))
                    push!(tensor_list, QR)
                end
            end
        
            # Define physical and virtual spaces
            Pspaces = [ProductSpace(physical_size, Real) for _ in 1:L]
            maxVspaces = [ProductSpace(physical_size, Real) for _ in 1:(L-1)]
        
            # Create FiniteMPS object
            return FiniteMPS(tensor_list, eltype(tensor_list); Pspaces=Pspaces, maxVspaces=maxVspaces)
        end
            QL_mid = QL * entanglement_spectrum
            QR_mid = entanglement_spectrum * QR

            push!(tensor_list, QL_mid)
            push!(tensor_list, QR_mid)
        end

        if i > Int(L/2)+1
            QR = transpose(TensorMap(randisometry((D*physical_size, D)), ℝ^(physical_size) ⊗ ℝ^(D), ℝ^(D)))
            push!(tensor_list, QR)
        end
    end

    # Define physical and virtual spaces
    Pspace = ProductSpace([Real for _ in 1:physical_size])
    Pspaces = [Pspace for _ in 1:L]
    maxVspace = ProductSpace([Real for _ in 1:physical_size])
    maxVspaces = [maxVspace for _ in 1:(L-1)]

    # Create FiniteMPS object
    return FiniteMPS(tensor_list, eltype(tensor_list); Pspaces=Pspaces, maxVspaces=maxVspaces)
end




entangledMPS = generate_entangled_mps(10, 5, 2)

# Once the MPS is generated, lets evaluate its entanglement properties along the chain
entanglement_along_chain = []
site_nr = collect(1:length(entangledMPS))
for i in site_nr
    S_exact =  sum(real, entropy(entangledMPS, i))
    push!(entanglement_along_chain, S_exact)
end

display(scatter(site_nr, entanglement_along_chain, xlabel="Site", ylabel="Entanglement entropy", title="Mixed canonical MPS optimized \n for high entanglement", yscale=:log10, legend=false, ylims = (0.01,1.0)))

randomMPS = FiniteMPS(rand, Float64,10, ℝ^2, ℝ^5)


#Once the MPS is generated, lets evaluate its entanglement properties along the chain
entanglement_along_chain = []
site_nr = collect(1:length(randomMPS))
for i in site_nr
    S_exact =  sum(real, entropy(randomMPS, i))
    push!(entanglement_along_chain, S_exact)
end

display(scatter(site_nr, entanglement_along_chain, xlabel="Site", ylabel="Entanglement entropy", title="Entanglement entropy in randomMPS",legend=false))
```
