```julia
using Plots

# extract data from the .jsons and plot with ticks to visualise the specifics
iterations = [0, 1, 2, 3]
num_boards = [0, 7200, 14400, 21600]
num_distinct_boards = [0, 4461, 8447, 12241]

# Create the line plot
p = plot(
    iterations, num_boards, 
    label = "Number of boards",
    xlabel = "Iteration",
    ylabel = "Experience Buffer Size",
    legend = :topleft, color=:blue
)

# Add the second line plot
plot!(p, iterations, num_distinct_boards, label = "Number of distinct boards", color=:orange)

# Add scatter plots
scatter!(p, iterations, num_boards, color=:blue, label="")
scatter!(p, iterations, num_distinct_boards,label="", color=:orange)

# Set the y-axis ticks
yticks_values = union(num_boards, num_distinct_boards)
yticks_labels = string.(yticks_values)  # Convert numbers to strings
yticks!(p, yticks_values, yticks_labels)

# Display the plot
display(p)
```
